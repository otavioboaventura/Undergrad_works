```python
# Biblioteca usadas
import numpy as np
import matplotlib.pyplot as pl
import sympy as sp
import scipy.linalg as la
import matplotlib.pyplot as pl

#############################################################################################################################
#                                                                                                                           # 
#                                             Absorvedor de masssas sintonizadas                                            #


###               SEÇÃO 1: Análise dos autovalores da matriz jacobina               ###

# Declaração das variáveis simbólicas
x1 = sp.Symbol('x1')
x2 = sp.Symbol('x2')
x3 = sp.Symbol('x3')
x4 = sp.Symbol('x4')
m1 = sp.Symbol('m1')
c1 = sp.Symbol('c1')
knl1 = sp.Symbol('knl1')
m2 = sp.Symbol('m2')
c2 = sp.Symbol('c2')
knl2 = sp.Symbol('knl2')

# Calculo da matriz Jacobina de forma simbólica
A = sp.Matrix([
                -(c1/m1)*x3 + (c2/m1)*(x4-x3) - (knl1/m1)*(x1**3) + (knl2/m1)*(x2 - x1)**3,
                (c2/m2)*(x4-x3) + (knl2/m2)*(x2 - x1)**3,
                x3,
                x4])

B = sp.Matrix([[x1, x2, x3, x4]])

J = A.jacobian(B)
print J

#-- Parâmetros

# Massa 1
J = J.subs(m1,1) # [Kg]
J = J.subs(c1,0.002) # [Ns/m]
J = J.subs(knl1,0.1)  # [N/m^3]
knl1 = knl1.subs(knl1,0.1)  # [N/m^3]

# TMD
J = J.subs(m2,0.05)      # [Kg]
J = J.subs(c2,0.002) # [Ns/m]
J = J.subs(knl2,0.1)  # [N/m^3]
knl2 = knl2.subs(knl2,0.1)  # [N/m^3]

# Condições inciais
J = J.subs(x1,-0.01)  # [m]
J = J.subs(x2,0)     # [m]
J = J.subs(x3,0)     # [m]
J = J.subs(x4,0)     # [m]
print J
                
# Calculo dos autovetores
autovalor = la.eigvals(J)*1000
print autovalor

# Classificação dos pontos de equilíbrio
pl.figure(1)
if ((autovalor[0] != autovalor[2]) & (autovalor[0]*autovalor[2]!=0) & (autovalor[0].imag==0) &
    (autovalor[2].imag==0)):
    if ((autovalor[0]*autovalor[2]>0) & (autovalor[0]<0) & (autovalor[2]<0)): # no hiperbólico assintoticamente estável
        pl.plot((knl1),(knl2),'ro') # estável 

    if ((autovalor[0]*autovalor[2]>0) & (autovalor[0]>0) & (autovalor[2]>0)): # Nó hiperbólico instável
        pl.plot((knl1),(knl2),'r*') #instável

    if (autovalor[0]*autovalor[2]<0): # sela hiperbolica instável
        pl.plot((knl1),(knl2),'rs') # instável

if (autovalor[0]*autovalor[2]==0): # Nulo - caso degenerados 
    pl.plot((knl1),(knl2),'go') # estavel

if ((autovalor[0].imag==0) & (autovalor[2].imag==0) & (abs(autovalor[0]) == abs(autovalor[2]))):

    if ((autovalor[0].real>0) & (autovalor[2].real>0)): # inflected node - instavel
        pl.plot((knl1),(knl2),'k*') # instavel

    else: # inflected node, assintoticamente estavel
        pl.plot((knl1),(knl2),'ko') # estavel

if  ((autovalor[0].imag)!=0 & (autovalor[2].imag!=0)): # complexo conjugado
    if ((autovalor[0].real==0) & (autovalor[2].real==0)):
        pl.plot((knl1),(knl2),'bo') # estável - centro eliptico 

    if ((autovalor[0].real!=0) & (autovalor[2].real!=0)):
        if ((autovalor[0].real>0) & (autovalor[2].real>0)):
            pl.plot((knl1),(knl2),'b*') # instável - foco hiperbolico

        else:
            pl.plot((knl1),(knl2),'bs') # estável - foco hiperbolico         
        
pl.ylabel(u'Rigidez secundária')
pl.xlabel(u'Rigidez principal')
pl.title(u'Diagrama de estabilidade da Estrutura')
pl.ylim(ymin = 0, ymax = 1)
pl.show()
```
